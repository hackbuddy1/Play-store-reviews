# Data Cleaning

```python
columns_to_drop = ['reviewCreatedVersion', 'repliedAt', 'replyContent']
```

```python
df.drop(columns=columns_to_drop, inplace=True, axis=1)
```

```python
df.sample(5)
```

```python
import re
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text
```

```python
df['content_clean'] = df['content'].apply(clean_text)
```

```python
df.head()
```

```python
from textblob import TextBlob
```

```python
def get_polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
df['polarity'] = df['content_clean'].apply(get_polarity)
```

```python
df.head()
```

```python
def get_sentiment_label(polarity_score):
    if polarity_score > 0:
        return 'Positive'
    elif polarity_score < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['sentiment'] = df['polarity'].apply(get_sentiment_label)
```

```python
df.head()
```

```python
df.sample(10)
```

# Visualization & Insights

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
sentimemt_counts = df['sentiment'].value_counts()
```

```python
print(sentimemt_counts)
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='sentiment', order=['Positive', 'Negative', 'Neutral'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

From the count-plot we can conclude that about 70% reviews are positive and the remaining 30% includes almost equal part of negative and neutral reviews.

It means, the core product is good. We don't need to change the whole app. But, we can also see 25% customers are not satisfied and we can't ignore them.

# Positive Reviews

```python
positive_text = " ".join(review for review in df[df['sentiment'] == 'Positive']
                         ['content_clean'] )
```

```python
wordcloud_positive = WordCloud(background_color='white', collocations=False).generate(positive_text)
```

```python
print("\n--- Word Cloud for Positive Reviews ---")
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud_positive, interpolation='bilinear')
plt.axis("off")  
plt.title('Most Common Words in Positive Reviews')
plt.show()
```

From the positive reviews we got that users consistently praise the app's simplicity and ease of use.

It's our Unique Selling Proposition(USP) that customer loves because of easy interface.

# Negative Reviews

```python
negative_text = " ".join(review for review in df[df['sentiment'] == 'Negative']['content_clean'])
```

```python
wordcloud_negative = WordCloud(background_color="black", colormap="Reds", collocations=False).generate(negative_text)
```

```python
print("\n--- Word Cloud for Negative Reviews ---")
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.axis("off")
plt.title('Most Common Words in Negative Reviews')
plt.show()
```

The primary source of customer dissatisfaction is technical performance, not missing features.

It means we have to work more on functionality of app not in new features. Our top priority is to keep the app stable and more fast.

1. Overall Sentiment

   - Around 70% of reviews are positive, which shows that users genuinely like the core product.

2. Strengths (Positive Reviews)

   - The biggest strength highlighted by users is the apps simplicity and ease of use.

   - This is your key value proposition  keep emphasizing it in marketing and product messaging.

3. Weaknesses (Negative Reviews)

   - The critical issues raised in negative reviews are not about   missing features, but about technical problems such as crashes and slow performance.

   - This indicates the product-market fit is strong, but the execution (stability and speed) needs improvement.
